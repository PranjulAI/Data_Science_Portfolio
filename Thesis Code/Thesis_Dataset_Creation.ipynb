**Importing various libraries and initialzing 4 lists**

```python
import selenium
from selenium import webdriver as wb
import time
cust_name=[] #list to store customer name
review_title=[] #list to store customer review title
review_text=[] #list to store customer review content
rating=[] #list to store customer rating
```

**Automating the opening of webpage**

```python
webD=wb.Chrome('chromedriver.exe')
webD.get('https://www.amazon.in/Bodyguard-Anti-Pollution-Face-Mask/product-reviews/B075XN3KML/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')
```

```python
time.sleep(15)
k=0 #loop control variable
n=28 #total no. of pages in the passed link
while (k<n):
    url=webD.current_url #stores the link of new page after each click
    webD.get(url) #opens the new link
    time.sleep(15)
    cm_cr_review_list=webD.find_element_by_id('cm_cr-review_list')
    #Stores customer names in the list
    names=cm_cr_review_list.find_elements_by_class_name('a-profile-name') 
    for i in range(0,len(names)):
        cust_name.append(names[i].text)
    
    #Stores review title in the list
    titles=cm_cr_review_list.find_elements_by_class_name('review-title-content')
    for j in range(0,len(titles)):
        review_title.append(titles[j].text)
    
    #Stores customer review content in the list
    content=cm_cr_review_list.find_elements_by_class_name('review-text-content')
    for l in range(0,len(content)):
        review_text.append(content[l].text)
    
    #Stores customer rating in the list
    rate=cm_cr_review_list.find_elements_by_css_selector('.review-rating span')
    for m in range(0,len(rate)):
        rating.append(rate[m].get_attribute('innerHTML'))
    
    #Controls the click should be done uptill second last page only
    if(k<n-1):
        link=webD.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
        link.click()
    k+=1
    
print("The total number of pages scraped were:",k)
```
